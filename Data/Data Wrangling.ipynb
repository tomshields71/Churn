```python
df = pd.read_csv(file)
df.head(10)
```

```python
def explore(header, d):
    """Prints size, columns, datatypes, info, description and percent of missing values for the given table.

    Keyword arguments:
    header -- the table name
    df -- the data frame
    """
    print(F'*****{header} info*****:')
    print("\nThe size is ")
    print(d.size)
    print('\nColumns')
    print(d.columns)
    print('"\nData types"')
    print(d.dtypes)
    print('')
    d.info(verbose = False);
    print()   
    print('\nGeneral Description of Numerical Data:')
    display(d.describe())
    print('\nLooking for empty values:')
    try:
        print(d.stb.missing(style=True))
    except:
        print(d.isna().sum())
    print('')
```

```python
explore('Churn Data Details', df)
```

# Data Wrangling

Total charges is currently an object and needs to be changed to a number.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors = 'coerce')
df.describe()
```

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
```

## Removing Null Values

```python
df.dropna(inplace = True)
```

```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
```

# Changing 'SeniorCitizen' to an object data type
It does not make sense to have this categorical column as the lone numeric data type.

```python
# Values changed from 1/0 to Yes/No. 
df["SeniorCitizen"]=df["SeniorCitizen"].astype(np.object)
df=df.replace({'SeniorCitizen': {1:'Yes', 
                                 0:'No' }})
```
